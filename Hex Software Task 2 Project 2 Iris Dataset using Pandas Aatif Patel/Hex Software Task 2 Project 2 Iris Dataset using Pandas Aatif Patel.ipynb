### 7. Data Aggregation with GroupBy: ###

For each species, calculate the average Petal Length and Petal Width. Then, identify which species has the highest average Petal Length-to-Width ratio.

```python
average_petal_measures = df.groupby('Species').agg({'PetalLengthCm': 'mean', 'PetalWidthCm': 'mean'})
average_petal_measures
```

```python
average_petal_measures['PetalLength_to_Width_Ratio'] = average_petal_measures['PetalLengthCm'] / average_petal_measures['PetalWidthCm']
average_petal_measures['PetalLength_to_Width_Ratio']
```

```python
highest_ratio_species = average_petal_measures['PetalLength_to_Width_Ratio'].idxmax()
highest_ratio_species
```

Q 1. How can I visualize the distribution of each feature (SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm) for the different species in the Iris dataset?

```python
df_melted = df.melt(id_vars="Species", var_name="Feature", value_name="Value")


plt.figure(figsize=(14, 10))
sns.boxplot(x='Feature', y='Value', hue='Species', data=df_melted)
plt.title('Feature Distribution by Species')
plt.show()
```

Q5. Create a heatmap to visualize the correlation between the features in the Iris dataset?

```python
# Drop non-numeric columns
numeric_df = df.select_dtypes(include=[float, int])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap of Feature Correlations in the Iris Dataset')
plt.show()
```

Q6. What is the most effective way to use a violin plot to compare the distribution of PetalWidthCm across species?

```python
sns.violinplot(x='Species', y='PetalWidthCm',hue='Species', data=df, inner='quartile', palette='Set2')
plt.title('Distribution of PetalWidthCm Across Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.show()
```

Q7. How can I visualize the class distribution (species) in the Iris dataset using a bar chart?

Q8 What are some ways to visualize the differences in feature distributions between the three species using density plots?

Solution 1

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(x='SepalLengthCm',data=df,hue='Species', fill=True, common_norm=False, palette='Set2')


plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.title('Density Plot of Sepal Length Across Species')

plt.show()
```

Solution 2

```python
g = sns.FacetGrid(df, hue="Species", height=4, aspect=1.5, palette='Set2')
g.map(sns.kdeplot, 'SepalLengthCm', fill=True, common_norm=False)
g.add_legend()
g.set_axis_labels('Sepal Length (cm)', 'Density')
g.fig.suptitle('Density Plot of Sepal Length Across Species', y=1.02)

plt.show()
```
