```python
df['Active_Cases']= df['Confirmed'] - df['Cured'] +df['Deaths']
```

```python
statewise = pd.pivot_table(df,values = ["Confirmed","Cured","Deaths"],index='State/UnionTerritory',aggfunc = max)
```

```python
statewise
```

```python
statewise["Recovery_Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Recovery_Rate"]
```

```python
statewise["Mortality_Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality_Rate"]
```

```python
statewise= statewise.sort_values(by = "Confirmed", ascending=False)
```

```python
statewise.style.background_gradient(cmap='cubehelix')
```

## Top 10 Active Cases State ##

```python
top_10=df.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by=['Active_Cases'],ascending=False).reset_index()
```

```python
fig = plt.figure(figsize=(20,10))

plt.title('Top 10 State With Most Active Cases In India',size = 30)

sns.barplot(data= top_10.iloc[:10],y='Active_Cases',x='State/UnionTerritory',linewidth=2, edgecolor = 'r')

plt.xlabel('Total Active Cases')
plt.ylabel('States')
```

## Top State With Highest Death ##

```python
top_10=df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()

fig = plt.figure(figsize=(20,10))

plt.title('Top 10 State With Most Deaths Cases In India',size = 30)

sns.barplot(data= top_10.iloc[:10],y='Deaths',x='State/UnionTerritory',linewidth=2, edgecolor = 'r')

plt.xlabel('Total Deaths Cases')
plt.ylabel('States')
```

## Growth Trends ##
