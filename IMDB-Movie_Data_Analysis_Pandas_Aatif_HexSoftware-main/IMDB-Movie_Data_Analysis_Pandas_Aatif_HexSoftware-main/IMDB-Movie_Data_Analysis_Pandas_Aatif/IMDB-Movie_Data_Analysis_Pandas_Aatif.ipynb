## Q12. Classify Movies Based on Ratings[Excellent,Good,Average] ##

```python
df.columns
```

```python
def rating(rating):
    if rating >= 7.0:
        return "Excellent"
    elif rating >= 6.0:
        return "Good"
    else:
        return "Average"
```

```python
df['Rating Category']=df['Rating'].apply(rating)
df['Rating Category']
```

## Q14. Find Unique Values From Genre ##

```python
df['Genre']
```

```python
list = []
for value in df['Genre']:
    list.append(value.split(','))
```

```python
list
```

```python
one_d=[]
for  item in list:
    for item in item:
        one_d.append(item)
```

```python
one_d
```

```python
unique = []
for item in one_d:
    if item not in unique:
        unique.append(item)
```
