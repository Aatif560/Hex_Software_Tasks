# Machine Learning Model #

## Load Data ##

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Lord\Downloads\delaney_solubility_with_descriptors.csv")
```

```python
df
```

## Data preparatiion ##

### Data Separations as x and y ###

```python
y = df['logS']
y
```

```python
x = df.drop('logS',axis=1)
```

```python
x
```

### Data Splitting ###

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train , y_test = train_test_split(x,y, test_size=0.2,random_state=100)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

## Building Model ##

### Linear Regression ###

```python
from sklearn.linear_model import LinearRegression
```

Training the model

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

Applying the model to make the prediction

```python
y_lr_train_pred = lr.predict(x_train)
```

```python
y_lr_test_pred = lr.predict(x_test)
```

```python
print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error , r2_score
```

```python
lr_train_mse =  mean_squared_error(y_train , y_lr_train_pred)
lr_train_r2 = r2_score(y_train,y_lr_train_pred)
```

```python
lr_test_mse =  mean_squared_error(y_test , y_lr_test_pred)
lr_test_r2 = r2_score(y_test,y_lr_test_pred)
```

```python
print("LR Mse(Train): ", lr_train_mse)
print("LR R2(Train): " ,lr_train_r2)
print("LR Mse(Test): " ,lr_test_mse)
print("LR R2(Test): " ,lr_test_r2)
```

```python
lr_results = pd.DataFrame(['LinearRegression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
```

```python
lr_results.columns = ['Method','Training MSE','Training R2','Test MES','Test R2']
```

```python
lr_results
```

## Random Forest ##

### Training the Model ###

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

Applying the model to make the prediction

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error , r2_score

rf_train_mse =  mean_squared_error(y_train , y_rf_train_pred)
rf_train_r2 = r2_score(y_train,y_rf_train_pred)
rf_test_mse =  mean_squared_error(y_test , y_rf_test_pred)
rf_test_r2 = r2_score(y_test,y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_results.columns = ['Method','Training MSE','Training R2','Test MES','Test R2']
```

```python
rf_results
```

## Model comparasion ##

```python
df_model = pd.concat((lr_results,rf_results), axis= 0)
```

```python
df_model
```
